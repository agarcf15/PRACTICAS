```python

csv = csv.dropna()
csv = csv.replace('\n', '. ', regex=True)
csv = csv.replace('"', '', regex=True)
csv = csv.replace('^\.', '', regex=True)
csv = csv.replace('\s\.', '', regex=True)
```

```python
csv
```

```python
columna = csv['Type of breach']
lista = columna.tolist()
tipos = set(lista)
count = len(tipos)
```

```python
func = lambda x: 0 if x == 'CARD' else 1 if x == 'INSD' else 2 if x == 'PHYS' else 3 if x == 'HACK' else 4 if x == 'DISC' else 5 if x == 'PORT' else 6 if x == 'STAT' else 7
csv['Label'] = csv['Type of breach'].apply(func)
```

```python
csv
```

```python
csv.to_csv('PRC_CLEAN.csv', index=False)
```

```python
func = lambda x: 'F' if x == 'CARD' else 'V' if x == 'INSD' else 'CI' if x == 'PHYS' else 'I' if x == 'HACK' else 'V' if x == 'DISC' else 'CI' if x == 'PORT' else 'CI' if x == 'STAT' else 'Others'
csv['Incibe'] = csv['Type of breach'].apply(func)
csv
```

```python
csv = csv.drop('Label', axis=1)
csv = csv.drop('Type of breach', axis=1)
```

```python
csv.to_csv('PRC_CLEAN.csv', index=False)
df = pd.read_csv('PRC_CLEAN.csv')

columnas_binarias = pd.get_dummies(df['Incibe'])
df = pd.concat([df, columnas_binarias], axis=1)
```

```python
df
```
